# Data Wrangling (Manipulacin de Datos)

```python
# Importacin de Libreras

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
```

## Lectura de los Datos

```python
# Lectura de datos desde una direccin URL

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
# Creacin de arreglo para los envabezados

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# Creacin del dataframe

df = pd.read_csv(filename, names = headers)
df.head()
```

## Identificacin y manejo de datos faltantes

```python
# reemplazo de "?" to NaN

df.replace("?", np.nan, inplace = True)
df.head(5)
```

### Evaluacin de los Datos faltantes

```python
missing_data = df.isnull()
missing_data.head(5)
```

### Contando los Valores faltantes en cada columna

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

### Lidiar con los Datos faltantes

#### Calculo del promedio para normalizar la columna

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)

avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

#### Reemplazo de los valores 'Nan' con el valor normalizado de la columna

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df["bore"].replace(np.nan, avg_bore, inplace=True)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

#### Reemplazo por el valor ms usado

```python
df['num-of-doors'].value_counts().idxmax()
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

#### Remover columans que tengan valores 'NaN'

```python
df.dropna(subset=["price"], axis=0, inplace=True)

# Reset de los ndices
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

## Correccin del formato de los Datos

### Listar el tipo de dato de cada columna

```python
df.dtypes
```

### Conversin de los Datos en el formato correcto

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
# Listado de los tipos de dato de cada columna despus de la conversin

df.dtypes
```

## Normalizacin de los Datos

### Conversin de Datos

```python
# Por ejemplo en este dataset conversin de unidades de km a mpg

df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

### Normalizacin de los Datos

```python
# Conversin a valores normalizados

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 

# mosntrar las columnas con los datos normalizados

df[["length","width","height"]].head()
```

# 'Binning' (transformacin de datos numeros a categricos discretos -bines-)

```python
# Conversin al tipo correcto

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# ploteo de un histograma de la columna del dato

plt.pyplot.hist(df["horsepower"])
```

```python
# determinacin de los bines

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# determinacin de los nombres de los grupos

group_names = ['Low', 'Medium', 'High']
```

```python
# uso de la funcin 'cut' para la determinacin de los bines

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
# ver cuntos registros tiene cada bin

df["horsepower-binned"].value_counts()
```

### Visualizacin de los bines

```python
plt.pyplot.hist(df["horsepower"], bins = 3)


plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
